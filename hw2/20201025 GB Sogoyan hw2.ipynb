````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
plt.style.use('fivethirtyeight')
````

````python
DATASET_PATH = 'creditcard.csv'
````

````python
df = pd.read_csv(DATASET_PATH)
df.head()
````

````python
df.info()
````

````python
df_class = df['Class'].value_counts()
df_class
````

````python
df_class.plot(kind='bar')
plt.show()
````

````python
df_class.plot(kind='bar', logy=True)
plt.show()
````

````python
df_v1_1 = df[df['Class'] == 1]['V1']
df_v1_0 = df[df['Class'] == 0]['V1']
````

#### Решение с помощью figures (2 гистограммы рядом)

````python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax1, ax2 = ax.flatten()

ax1.hist(df_v1_1, bins=20, density=True, alpha=0.5, label='Class 1', color='red')
ax1.set_xlabel('Class')

ax2.hist(df_v1_0, bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
ax2.set_xlabel('Class')

legend1 = ax1.legend(frameon=False)
legend2 = ax2.legend(frameon=False)
````

#### 2 гистограммы на одном графике

````python
plt.hist(df_v1_1, bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.hist(df_v1_0, bins=20, density=True, alpha=0.5, label='Class 0', color='grey')

legend = plt.legend(loc='best', frameon=False)
plt.setp(legend.get_texts(), color='grey')
plt.xlabel('Class')

plt.show()
````
